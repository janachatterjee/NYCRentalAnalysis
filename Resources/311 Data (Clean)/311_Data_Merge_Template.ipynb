# 311 Clean CSV Merge Template

```python
# Dependencies

import os
import csv
import pandas as pd
```

```python
# Store each CSV file into a variable

bronx_csv = "Resources/311BX2018Data.csv"
brooklyn_csv = "Resources/311BK2018Data.csv"
manhattan_csv = "Resources/311MN2018Data.csv"
queens_csv = "Resources/311QN2018Data.csv"
statenisland_csv = "Resources/311SI2018Data.csv"
```

```python
# Next, read each CSV into a Pandas DataFrame

bx_df = pd.read_csv(bronx_csv)
bk_df = pd.read_csv(brooklyn_csv)
manhat_df = pd.read_csv(manhattan_csv)
qz_df = pd.read_csv(queens_csv)
si_df = pd.read_csv(statenisland_csv)
```

```python
# K, now merge all the dataframes based on the dates they share
letsjoin_df = bx_df.append(bk_df).append(manhat_df).append(qz_df).append(si_df)
letsjoin_df
```

```python
# Yay! It worked! Now, let's write the info into a new CSV
letsjoin_df.to_csv("Merged3112018Data.csv", encoding="utf-8", index=False)
```
