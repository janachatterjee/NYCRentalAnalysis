```python
# Dependencies
import requests
from pprint import pprint
from config import api_key

url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"
```

```python
# Define a query search that will look for AirBnB articles, then StreetEasy & 311
# query = "AirBNB"
# query = "StreetEasy"
query = "311"
# query = "Complaints"
```

```python
# Next, build the query URL
query_url = url + "api-key=" + api_key + "&q=" + query
```

```python
# Alrighty, let's request the articles from the NYT API
articles = requests.get(query_url).json()

# Use list comprehension so the "response" property in articles contains the actual articles
articles_list = [article for article in articles["response"]["docs"]]
pprint(articles_list)
```

```python
 # Print the web_url of each stored article
print("Here are the articles you requested")
for article in articles_list:
    print(article["web_url"])
```
